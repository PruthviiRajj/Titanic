```python
import zipfile
zip_ref = zipfile.ZipFile('titanic.zip', 'r')
zip_ref.extractall('/content')
zip_ref.close()
```

```python
import pandas as pd
import numpy as np
```

```python
train = pd.read_csv('/content/train.csv')
test = pd.read_csv('/content/test.csv')
```

```python
train.info()
```

```python
train.drop(['Name','Fare','Cabin','Ticket'],axis=1,inplace=True)
test.drop(['Name','Fare','Cabin','Ticket'],axis=1,inplace=True)
```

```python
train.drop(['PassengerId'],axis=1,inplace=True)
```

```python
train.head()
```

```python
train['Age'].mean()
```

```python
train['Age'].fillna(train['Age'].mean(),inplace=True)
```

```python
test['Age'].fillna(test['Age'].mean(),inplace=True)
```

```python
Passengerid  = test['PassengerId']
```

```python
test.drop(['PassengerId'],axis=1,inplace=True)
```

```python
test.info()
```

```python
train.head()
```

```python
train['Parch'].value_counts()
```

```python
train['SibSp'].value_counts()
```

```python
train['Family_size'] = train['SibSp'] + train['Parch'] + 1
test['Family_size'] = test['SibSp'] + test['Parch'] + 1
```

```python
train.drop(['SibSp','Parch'],axis=1,inplace=True)
test.drop(['SibSp','Parch'],axis=1,inplace=True)
```

```python
train['Family_size'].value_counts()
```

```python
def family(Family_size):
  if Family_size == 1:
    return 'Alone'
  if Family_size > 1 and Family_size < 5:
    return 'Medium'
  else:
    return 'Large'
```

```python
train['Embarked'].fillna('S',inplace=True)
test['Embarked'].fillna('S',inplace=True)
```

```python
train.isnull().sum()
```

```python
train['family'] = train['Family_size'].apply(family)
test['family'] = test['Family_size'].apply(family)
```

```python
train.head()
```

```python
train.drop(['Family_size'],axis=1,inplace=True)
test.drop(['Family_size'],axis=1,inplace=True)
```

```python
train.head()
```

```python
passengerid = Passengerid.values
```

```python
train = pd.get_dummies(train,columns=['Pclass','Sex','Embarked','family'],drop_first=True)
```

```python
test = pd.get_dummies(test,columns=['Pclass','Sex','Embarked','family'],drop_first=True)
```

```python
train
```

```python
X = train.iloc[:,1:].values
y = train['Survived'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2, random_state=225)
```

```python
X_train.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
classifier = DecisionTreeClassifier()
```

```python
classifier.fit(X_train,y_train)
```

```python
y_predict = classifier.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_predict,y_test)
```

```python
Xf = test.iloc[:,:].values
```

```python
y_final = classifier.predict(Xf)
```

```python
final = pd.DataFrame()
```

```python
final['PassengerId'] = Passengerid
final['Survived'] = y_final
```

```python
final
```

```python
final.to_csv('final.csv',index=False)
```
